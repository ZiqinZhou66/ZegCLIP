```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
outputs = np.load(
    "/mnt/Enterprise2/safal/ZegCLIP/pretrained/outputs/kvasir.pkl",
    allow_pickle=True,
)
gt = np.load(
    "/mnt/Enterprise2/safal/ZegCLIP/pretrained/outputs/kvasir_gt.pkl",
    allow_pickle=True,
)
```

```python
rand_idx = [np.random.randint(0, len(outputs)) for _ in range(10)]
```

```python
# create a grid of 2x5 images
fig, ax = plt.subplots(2, 5, figsize=(20, 8))
for i in range(2):
    for j in range(5):
        ax[i, j].imshow(outputs[rand_idx[i * 5 + j]], cmap="gray")
        ax[i, j].axis("off")
```

```python
# create a grid of 2x5 images
fig, ax = plt.subplots(2, 5, figsize=(20, 8))
for i in range(2):
    for j in range(5):
        ax[i, j].imshow(gt[rand_idx[i * 5 + j]], cmap="gray")
        ax[i, j].axis("off")
```
